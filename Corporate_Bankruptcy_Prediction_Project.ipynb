## Variance Threshold, should we use this?

Recall that the sample mean formula is:

$$\bar x = \frac{1}{n}\sum^n_{i=1} x_i$$


Recall that the sample variance formula is:

$$Var(x_i) = s^2 = \frac{1}{n-1} \sum^n_{i=1} (x_i - \bar x)^2$$

Now, lets talk about a scenario where variance is good, and variance is bad.

---

### Variance in the data: GOOD

Lets try to select predictor variables that have some amount of variance

You may be thinking: "wait isn't variance bad?"

NO! In terms of data, variance is GOOD! Variance gives information!

If our data only has 40 students in a class and they all have blue eyes, that is not a good predictor variable! It will not give any information on what nationality each student is, what their heights are, what their personalities are, and so on!

Recall that the defintion of variance is how spread the data is! The more spread, the more information! This is a reason why PCA measures variance!


---

### Variance in the model: BAD

Variance being "bad" will depend on the context.

One instance of when variance is bad is when we want good predictions. If our machine learning model predictions has high variance, then our machine learning model is doing a terrible job at predicting values. This is because it's predictions are all over the place!

---

What predictors have a variance of at least 0.01?


```python
from sklearn.feature_selection import VarianceThreshold

# threshold = 0.01 -> any predictors that have a variance of 0.01 or lower
# will not be selected

selector_function = VarianceThreshold(threshold = 0.01)


X = company_df.drop("Y", axis = 1)
#print(X)
#print(X.columns)

X_small = selector_function.fit_transform(X)

# We need to get our new predictors!
new_columns_boolean = selector_function.get_support()
print(new_columns_boolean)

# filter out the important predictors:
important_predictors = X.columns[new_columns_boolean]
print(important_predictors)

X_small = pd.DataFrame(X_small, columns = important_predictors )
#print(X_new)


X_small

```

## What predictor variables are we selecting?

Overall, I will actually keep all predictors for my machine learning models. Let me explain.

Firstly, we need to know about the "Curse of Dimensionality. A GeeksForGeeks Website states that: "Curse of Dimensionality in Machine Learning arises when working with high-dimensional data (data with a lot predictors), leading to increased computational complexity, overfitting, and spurious correlations" (jayantiks24).

There are some situations where data sets have significantly more predictor variables ($p$) then observations ($n$): $p >> n$.

For instance, someone could use 500,000 single nucleotide polymorphisms to predict someone's blood pressure. Lets say we have 200 patients. Then $n = 200$ and $p = 500,000$ (James, 238-239).

If some data sets have half a million predictors, then we can easily handle 95 predictors!

Models such as LASSO and Ridge can help account for the curse of dimensionality problem. They will penalize big coefficients and this is a attempt to get the least error possible.  

Models such as random forests and XGBoost are not affected by the amount of predictors. They can be tuned so that each tree only uses some predictors, not all 95.

## Unsupervised Model 1: Principal Components Analysis (PCA)

Principal Components Analysis (PCA), is a dimension reduction technique that reduces $p$ number of dimensions/predictors into $k$ amount of dimensions. Where: $k < p$

New variables called Principal Components (PCs) are created. These PCs contain linear combinations of all the predictor variables.

One main reason we use PCA is to visualize the data in a 2D or 3D space. Furthermore, we may find any patterns in this space.


Lets start coding our PCA model!



```python
# ALERT: We want to first standardize the data before doing PCA!

from sklearn.preprocessing import StandardScaler

scaler_function = StandardScaler()

X_scaled = scaler_function.fit_transform(X)
# X_scaled
```

### How many PCs do we select?

```python
from sklearn.decomposition import PCA


pca_model_scree = PCA().fit(X_scaled)


plt.figure(figsize=(15,8))  # width= 15 inches, height= 8 inches

ax = plt.axes()


cumulative_var = np.cumsum(pca_model_scree.explained_variance_ratio_)


ax.plot(np.cumsum(pca_model_scree.explained_variance_ratio_),
        marker = "o",color = "green")
# we need to set a marker

ax.axhline(y = 0.9, color = "blue", linestyle = "--", linewidth = 2,
           label = "the 0.9 cumulative variance marker" # this can be used for the legend
           )

ax.set_title("Scree plot of PCs. How much should we select?", fontsize = 20)
ax.set_xlabel("PCs / number of components")
ax.set_ylabel("Cumulative variance explained")

# set the ticks from 0 to 95
ax.set_xticks(range(0, len(cumulative_var), 5)) # only show increments of 5

# set the tick labels to be from 1 to 95
ax.set_xticklabels(range(1,len(cumulative_var)+1, 5));

ax.legend();


```

Ouch. 2 components explain about 21% of the total variance, while 3 components only explain about 25% of the total variance.

This is not ideal but we need to only choose 2 and 3 PCs to visualize the data.

If we really want to pick the number of PCs that explain a significant amount of the variance, we should pick 45 PCs, since 45 PCs explains 90% of the variance.


```python
from sklearn.decomposition import PCA # 1. fetch class

#2. intialize model

pca_model = PCA(n_components = 3, random_state=1)

# 3. fit model (and transform if needed)

X_pca = pca_model.fit_transform(X_scaled)

print(X_pca)

# this above is a numpy array, we need to make it a data frame!

# make sure we have column names in this data frame!
pca_df = pd.DataFrame(X_pca, columns= ["PC1","PC2","PC3"])
#pca_df


pca_df = pd.concat([pca_df, company_df["Y"]], axis = 1)
#pca_df

# lets make a categorical labels:

pca_df["bankrupt_label"] = pca_df["Y"].map({0:"Not Bankrupt",1:"Bankrupt"})
pca_df
```

## Unsupervised Model 2: Locally Linearly Embedding (LLE)

LLE is another dimension reduction technique just like PCA. LLE has one advantage, it can effectively do dimension reduction on non-linear data.


This is because LLE only  calculates pairwise distances of nearby points.

* For LLE, it would just calculate pairwise distances of the 100 nearest points. (We can adjust the amount of points by using "n_neighbors = " if we think 100 is not good.) Overall, LLE only considers local structure of the data.

For a in depth explenation of LLE, please reference my Pokemon MEGA Project.


```python
from sklearn.manifold import LocallyLinearEmbedding

# make a 2D LLE plot

lle_model = LocallyLinearEmbedding(n_neighbors = 100, n_components = 2,
                                   method = "modified", eigen_solver = "dense")

X_lle = lle_model.fit_transform(X)
print(X_lle)


lle_df = pd.DataFrame(X_lle, columns = ["Component 1", "Component 2"])
print(lle_df)




```

## Cross Validation (CV)

Having one training set and one testing set is a good idea in machine learning. But can we do better?

YES! We can do better! Cross Validation (CV) is a process where we can create multiple sets, and these sets can be the training set or testing set. A popular CV style is the K-fold Cross validation.

K-fold cross validation is the process of having a $K$ amount of sets. We make $K-1$ the training set, and the left over set is the testing set. Here is a example:

---
### CV example:

* Lets say we have 500 observations. It is recommended to shuffle these observations due to data entry reasons.

For instance, the first 100 observations may only have point guard NBA players, then the second 100 obervations only has center NBA players, and so on.

Point guards tend to have a smaller height and have a lower age compared to Centers, so having one set just have point guards does not seem like a good way to train a model.

We would want to randomize these players to make each set have diverse information. We want diverse information so that a model can get "trained better".

* Lets make 5 sets. This causes each set to have 100 observations.

* K-Fold CV steps:

1. Make the first set the testing set. The second to fifth set will become the training set. Train the model using the training set and use the testing set values to make predictions. Calculate our error or scores for this first testing set. This will be $E_1$. (For regression, we usually use Mean Squared Error. For classification, we could use accuracy score, F1 score, roc auc scores, and more)

2. Make the second set the testing set. The first and third to fifth set will become the training set. Train the model using the training set and use the testing set values to make predictions. Calculate our error or scores for this first testing set. This will be $E_2$.

3. I think you see the pattern. We will eventually have 5 errors or scores. We then usually take the mean of these errors or scores. This mean is a accurate estimate of how well our model does at predicting the response values.


CV mean score Formula:

$$CV ~ Mean ~ score  = \frac{1}{n} \sum^n_{k=1} S_k$$

## PR and AUC

Data Camp's Tutorial on PR and AUC states that: "A precision-recall curve helps you decide a threshold on the basis of the desirable values of precision and recall. It also comes in handy to compare different model performance by computing 'Area Under the Precision-Recall Curve' abbreviated as AUC" (Chugh).



---

### **PR** = **P**recision **R**ecall






The Precision Recall curve (PR) is just a graph that shows how well a classification model seperates positive cases from a negative cases. All thresholds are considered. For instance, if we have a threshold of 0.5, then any probablity of being a positive class that is greater than 0.5 becomes classified as a positive. Different thresholds such as 0.2 and 0.7 will be tested as well.


Here are the notations and formulas:

* $TN$ = True Negative
* $FN$ = False Negative
* $FP$ = False Positive
* $TP$ = True Positive


1. Measuring Precision

* Precision is the portion of total positive predictions that are correct.

* In other words, how often do we classify bankrupt companies correctly? A high precision means that model does not make many errors on classifying a true bankrupt company.

$$Precision = \frac{TP}{TP + FP}$$


2. Measuring Recall.
$$Recall = \frac{TP}{TP + FN}$$

* Recall is the portion of true positives divided by the total amount of actual positives.

* In other words, does this model find most bankrupt companies? A high recall means the model catches the most bankrupt companies available, even if it means misclassifying some companies.

In the PR graph, we have:

* y-axis = Precision (P)

* x-axis = Recall (R)



---
### **AUC** = **A**rea **U**nder the PR **C**urve

The Area Under the PR Curve is the area number that represents how well the classification model classifies the two classes.


You might think "oh wow, curves and area under the curves, this sounds like calculus!" And you would be right! We are essentially integrating the PR curve, and the area is calculated with this integral:

$$AUC = \int^{R = 1}_{R = 0} \text{height dx} = \int^1_0 P(R)  d(R)$$


* Recall that the y-axis has P.  

* Recall that the x-axis has R, and in single variable calculus, we usually integrate with respect to (wrt) x. Therefore we integrate wrt R.

Furthermore, since the integration bounds uses x values, R values go in the integration bounds.

* You can essentially think: $y = P$, $x = R$

* Furthermore, P represents the height and R represents the width. Integrate these to get our area.

As a example, lets integrate $f(x) = 4x$ over domain $[0,3]$:

$$\int^3_0 f(x) dx = \int^3_0 4x dx = 2x^2 |^3_0 = 2(3)^2 - 0 = 18 - 0 = 18$$

* $4x$ is the height, $dx$ is the width.


### How to interpret AUC considering PR curves:

So what AUC numbers represent a model classifying poorly or fantatsically??

Well AUC can integrate between 0 to 1. Here are the interpretations of some numbers.

* Our baseline number is 0.03, since about 3 percent of the companies are bankrupt:

* $0 \le AUC < 0.03$ -> terrible.

* $0.03 \le AUC \le 0.04 $ -> random performance.  

* $0.04 < AUC \le 0.10$ -> poor.

* $0.10 < AUC \le 0.20$ -> fair.

* $0.20 < AUC \le 0.30$ -> good.

* $0.30 < AUC \le 0.50$ -> great.

* $0.50 < AUC \le 0.80$ -> fantastic!

* $0.80 < AUC \le 1.00$ -> AMAZING!


* Please note that the AUC interpretations will depend on the data set!


---
### What if we have multiple classes?

You may notice that ROC and AUC mainly uses two classes, the positive and negative classes. So can we use ROC and AUC with multiple classes?

YES! We can treat the class of interest as the positive class, and the other classes as the negative class!

## AUC score is what matters most! Not ROC!

When people talk about ROC and AUC, AUC is what gives our metric on if a model performed greatly or not! ROC is only a curve, nothing else!

Now, the curve that ROC makes is still important since it helps us make our AUC score in the first place.

Ultimately, AUC gives a number gives us a quantifiable measurement. (Just like accuracy scores and F1 scores.)

Lets make a function that explains how well out AUC score is!


```python
# AUC check function:

def auc_check(cv_number):
  if 0 <= cv_number < 0.03:
    print("The model did a terrible job in predicting the bankrupt companies.")
  elif 0.03 <= cv_number <= 0.04:
    print("The model had a random performance in predicting the bankrupt companies.")
  elif 0.04 < cv_number <= 0.10:
    print("The model did a poor job in predicting the bankrupt companies.")
  elif 0.10 < cv_number <= 0.20:
    print("The model did a fair job in predicting the bankrupt companies.")
  elif 0.20 < cv_number <= 0.30:
    print("The model did a good job in predicting the bankrupt companies.")
  elif 0.30 < cv_number <= 0.50:
    print("The model did a great job in predicting the bankrupt companies.")
  elif 0.50 < cv_number <= 0.80:
    print("The model did a fantastic job in predicting the bankrupt companies!")
  elif 0.80 < cv_number <= 1.00:
    print("The model did a AMAZING job in predicting the bankrupt companies!!!")
  else:
    print("Was the mean CV number between 0 or 1?")



```

## Classification Model 2: Logistic Ridge


Logistic Ridge is a special type of logistic regression where the beta coeffcients shrink towards zero, but never become zero.

The reason we use Logistic Ridge is to improve the PR-AUC score! Lower beta coefficients can reduce overfitting which can increase the PR-AUC score when using testing data!

We normally use Logistic Ridge if they believe the predictors are correlated.

* The first 3 predictors are already suspected of being correlated because they are about "Return on Total Assets (ROA):

X1 - ROA(C) before interest and depreciation before interest: Return On Total Assets(C)

X2 - ROA(A) before interest and % after tax: Return On Total Assets(A)

X3 - ROA(B) before interest and depreciation after tax: Return On Total Assets(B)

Logistic Ridge will shrink these beta coefficients for X1, X2, and X3 since these predictors are correlated.

---

### Theory of Logistic Ridge:


* In logistic LASSO, a lambda($\lambda$) penalty term is used.  The bigger the lambda penalty, the more the beta coefficients shrink towards 0. BUT for Logistic Ridge, the beta coefficients will only be close to 0, not exactly 0!  

* The L2 norm: $||\beta||_2 = \sum^p_{j =1}  \widehat \beta_j^2$ is the reason why our beta coeffcients can become close to zero, but not exactly zero! Think of a 2D plot where the the range of values for x and y are [-1,1]. With L2 norm, we can only make circular lines, and due to the [-1,1] range, we end up making a circle shape. The circle has 4 reference points: (1,0),(0,1), (-1,0), (0,-1).


For more information on how Logistic Regression works, please refer to the Pokemon MEGA Project.


---
### ALERT:
Python's scikit-learn uses the parameter of $\alpha$ instead of $\lambda$. So please keep this in mind!





You will notice in the code below, we dont have a ($\alpha$) value. Instead we have a C value, which is just the inverse alpha.

$$C = \frac{1}{\alpha}$$  


```python
from sklearn.linear_model import LogisticRegressionCV


ridge_model = LogisticRegressionCV(
    Cs=7,           # list of log spaced C's to try EX: C's = [0.001,0.01,0.1,1,10,100,]
    cv=3,            # 3-fold cross-validation
    penalty="l2",    # L2 = Ridge
    scoring="average_precision", # PR-AUC
    max_iter=1000,   # number of steps needed for convergence to find a good C
    random_state = 1
    )


cv_scores = cross_val_score(ridge_model, X_train, y_train,
                            cv = sk_folds, scoring = "average_precision")


cv_scores

```

```python
# checking the AUC scores:

mean_cv_ridge = round(np.mean(cv_scores),3)

print("The mean AUC score is:", mean_cv_ridge)
auc_check(mean_cv_ridge)

```

### Conclusion for Logistic Ridge

Well, that was surprising. Logistic Ridge had a lower PR-AUC score than regular Logistic Regression.

Perhaps Logistic Ridge lowered the variance, but the increase in bias overpowered the variance reduction.

---
The above statement is inspired by the bias-variance trade off. We want a model that has the right amount of bias and variance so that the model has the least amount of error. (Or conversely, the highest score).

A model has lower variance if it is less complicated. AKA it has fewer predictors, the coefficients are weaker, there are less model parameters, etc.

A model has lower bias if it is more complicated. AKA it has more predictors, the coefficients may be stronger, there are more parameters, and more. (The risk with less bias in the training set is that the model will overfit it too much and give terrible predictions when using the test set.)




## Classification Model 3: Logistic LASSO  

Logistic LASSO is a special type of logistic regression where the beta coeffcients shrink towards zero, and CAN become zero. This means Logistic  LASSO acts as a variable selector. Lets say $\hat \beta_1 = 0$, then the Base_stats predictor of $x_1$ does nothing, since $0 * any~number = 0$.

* The first 3 predictors are already suspected of being correlated because they are about "Return on Total Assets (ROA):

X1 - ROA(C) before interest and depreciation before interest: Return On Total Assets(C)

X2 - ROA(A) before interest and % after tax: Return On Total Assets(A)

X3 - ROA(B) before interest and depreciation after tax: Return On Total Assets(B)

Logistic LASSO could shrink one or two of the beta coefficients for X1, X2, and X3 to zero since only one of them is needed.

For instance, if LASSO causes $\hat \beta_3 = 0$, then it does not matter what value X3 is, $0 * value = 0$

---

### Theory of Logistic LASSO:

* In logistic LASSO, a lambda($\lambda$) penalty term is used.  The bigger the lambda penalty, the more the beta coefficients shrink towards 0. If the lambda penalty becomes big enough, some beta coefficients will be effectively 0.


* The L1 norm: $||\beta||_1 = \sum^p_{j =1}|  \widehat \beta_j|$ is the reason why our beta coeffcients can become zero! Think of a 2D plot where the the range of values for x and y are [-1,1]. With L1 norm, we can only make straight lines, and due to the [-1,1] range, we end up making a diamond shape. There are 4 sharp axis points of the diamond: (1,0),(0,1), (-1,0), (0,-1). If the RSS countour touches one of those axis points is bound to be a x value ($\widehat \beta_1$) or a y value ($\widehat \beta_2$) that has a value equal to zero.


Again, please check the Pokemon MEGA Project for more information on Logistic LASSO.

---

$$C = \frac{1}{\alpha}$$

```python
from sklearn.ensemble import RandomForestClassifier


# What each parameter does:

# n_estimators = 500 -> number of trees, 500 trees in this case

# oob_score=True -> gives out-of-bag error


# max_depth = 3 -> 3 is the max layers of a tree

# min_samples_split = 10 -> needs at least 10 observations in a node for a split
# a split makes a branch, no split means no branch is made

# min_samples_leaf = 5 -> a leaf needs at least 5 observations.
# The leaf is the end part of a tree, where we get our predictions

# max_features = "sqrt" -> each tree only uses the square root amount of the
# total predictors. Here, we got 4 predictors, so sqrt(4) = 2,
# each tree uses 2 predictors, and these 2 predictors are selected at random
# ALERT: we usually USE this since we want to make each tree totally different!

# class_weight = "balanced" -> handles class imbalance, since we have
# less legendaries random forest tries to give a higher weight to legendaries.
# higher weights means the models tries pay more attention to it

# bootstrap = True by default, don't need to specify this!

# random_state = 1 -> set.seed(1) in R, always make the same tree

# all these parameters above can reduce overfitting and hopefully gives less error
# random_state does NOT contribute to reducing overfitting


rf_model = RandomForestClassifier(n_estimators = 500,
                                  oob_score = True,
                                  max_depth = 3,
                                  min_samples_split = 10,
                                  min_samples_leaf = 5,
                                  max_features = "sqrt",
                                  class_weight = "balanced",
                                  random_state = 1)


cv_scores = cross_val_score(rf_model, X_train, y_train,
                            cv = sk_folds, scoring = "average_precision" )


cv_scores
```

## Classification Model 5: XGBoost  

* XGBoost stands for Extreme Gradient Boosting. For classification, it sums the  baseline number and predictions from multiple trees to  produce one number. Next, it converts that number to a probability.


* Random Forest is a parallel approach since it uses independent trees. When training  the trees, they are made randomly due to the bootstrap process (I explain the bootstrap process in great detail in my Pokemon MEGA Project!).

* XGBoost is a sequential approach. The trees are trained by each tree predicting then learning: We start with a baseline value. Next, tree 1 predicts a residual, then tree 2 learns from tree 1 and the baseline value, and tree 2 gives its own residual. Next, tree 3 learns from tree 2, tree 1, and the residual, and tree 3 gives its own predicition. Repeat this for M trees. Combine all tree values plus the baseline value to get our final prediction for observation. Repeat for observarions 2 to n.    


For more information on the theory and the  mechanics of XGBoost, please reference my Pokemon MEGA Project.  


---
### NOTE on using CV for Random Forest parameters

I am aware that we can use CV to find the best parameters (such as n_estimators) that can give the best predictions for random forest.

But this is too computationally expensive. This especially true when finding CV estimates for the PR-AUC scores is already computationally expensive.  




```python
from xgboost import XGBClassifier



# XGBoost parameters:

# 1. objective = "reg:squarederror" -> use MSE
# this is the default, so no need to mention this!

# 2. n_estimators -> amount of trees to use, called boosting rounds for XGBoost
# n_estimators = 500 trees

# 3. learning_rate (eta) -> how much each tree learns from previous tree
# learning_rate = 0.05 -> eta = 0.05

# 4. max_depth -> how complex each tree is. depth = 1 -> 2 branches.
# depth = 2 -> at most 4 branches, depth = 3 -> at most 8 branches

# 5. subsample -> fraction of training rows used for each tree, prevents overfitting
# subsample = 0.8 -> only use 80% of training data for each tree

# 6. colsample_bytree -> fraction of total predictors to use for each tree
# similar to max_features in random forest
# colsample_bytree = 0.8 -> use 80% of predictors, in our case we have 4 so use 3.

# 7. gamma -> gamma = 0 makes big complex trees, gamma = 5 makes small simple trees

# 8. min_child_weight -> number of samples or weights needed to make a split.
# similar to min_sample_split in random forest, BUT if we use weights, can be different
# min_child_weight = 5 -> need at least 5 pokemon to make a split
# for instance, if we give a legendary a weight of 5, and regular pokemon a weight of 1,
# 2 legendary pokemon and 5 regular pokemon -> (2*5) + 5 = 15 as weight.

# 9. scale_pos_weight -> do number_majority_class / number_rare_class
# this accounts for class imbalance! We have a lot more non legendaries than legendaries
# NOTE: This only works for binary classification!

# 10. random_state = 1 -> set the seed as 1, this makes results reproducible.
# think of this as a minecraft seed, the same minecraft seed always gives a same world!


# lets get our non bankrupt and bankrupt companies:

non_bankrupt_num = sum(y_train == 0)
print("Non Bankrupt companies: ", non_bankrupt_num)

bankrupt_num = sum(y_train == 1)
print("Bankrupt companies: ", bankrupt_num)








xgboost_model = XGBClassifier(n_estimators = 500,
                          learning_rate = 0.05, # eta
                          max_depth = 3,
                          subsample = 0.8,
                          colsample_bytree = 0.8,
                          gamma = 3,
                          min_child_weight = 10,
                          scale_pos_weight = (non_bankrupt_num/bankrupt_num),
                          random_state = 1
                          )

cv_scores = cross_val_score(xgboost_model, X_train, y_train,
                            cv = sk_folds, scoring = "average_precision")

cv_scores
```

```python
mean_cv_xgboost = round(np.mean(cv_scores),3)

print("The mean AUC score is:", mean_cv_xgboost)
auc_check(mean_cv_xgboost)


```

Recall in our 2D and 3D PCA plots that most bankrupt companies tend to be grouped in one area. What if we use those PCs to find out which companies are bankrupt or not? This might be a smart idea considering that the PC's avoid the "Curse of Dimensionality" problem that we talked about earlier in the Project. Using 3 PCs is more easy to use utilize than all 95 predictors.


* This model is simple. We select a testing set point to classify. Then look at the K nearest training set points and take the majority vote of the classes.

* We usually select the nearest points by the shortest euclidian distance possible. Think of the distance formula. Here is a example with three variables: $d_E = \sqrt{(x_0 - x_1)^2 + (y_0 - y_1)^2 + (z_0 - z_1)^2}$

* For example, say we select $K = 5$. In this point, what the class labels of its 5 nearest points? Lets say the 4 other points are not bankrupt company, but 5th point is a bankrupt company. Then by majority vote of 4 to 1, we classify the selected point as a nonbankrupt company.





```python
from sklearn.neighbors import KNeighborsClassifier

# NOTE: no need to standardize since PCA was already scaled!

cv_mean_list = []

# FOR loop for multiple K values ->

for guys in range(1,21):

  knn_model = KNeighborsClassifier(n_neighbors = guys)

  cv_scores = cross_val_score(knn_model, X_train, y_train,
                              cv = sk_folds, scoring = "average_precision")

  cv_mean = round(np.mean(cv_scores),3)

  cv_mean_list.append(cv_mean)



print(cv_mean_list)

print()

pr_max_value = np.max(cv_mean_list)
print("What is the max PR-AUC score:", pr_max_value) # y value

pr_max_location = np.argmax(cv_mean_list) + 1 # need to do + 1 to account for 0 index
print("How many neighbors gives the max PR-AUC score:", pr_max_location) # x value

# assign pr_max_value as the best cv_mean value

mean_cv_knn = pr_max_value
```

### What K neighbors gives the highest PR-AUC score?

```python

fig = plt.figure(figsize = (15,8))

plt.scatter(range(1,21), cv_mean_list, s = 100,
            label = " K Neighbors score plot")

plt.plot(range(1,21), cv_mean_list, linewidth = 2)

plt.scatter(pr_max_location, pr_max_value, s = 300,
            edgecolors = "red", facecolors = "none",
            label = "K = 14 neighbors gives the highest PR-AUC score.")


plt.title("What K amount of neighbors gives the highest PR-AUC score?", fontsize = 25)

plt.ylabel("PR-AUC score")
plt.xlabel("Amount of K neighbors")

plt.legend();

```

### Conclusion for KNN

It appears that 14 neighbors gives a PR-AUC score of 0.093, but this is way lower than XGBoost's PR-AUC score of 0.477.

# üèÜ Final Classification Model

XGBoost is the clear winner, so lets use all the of the data from the training set and compare it against the testing set data.



```python
from xgboost import XGBClassifier

# XGBoost parameters:

# 1. objective = "reg:squarederror" -> use MSE
# this is the default, so no need to mention this!

# 2. n_estimators -> amount of trees to use, called boosting rounds for XGBoost
# n_estimators = 500 trees

# 3. learning_rate (eta) -> how much each tree learns from previous tree
# learning_rate = 0.05 -> eta = 0.05

# 4. max_depth -> how complex each tree is. depth = 1 -> 2 branches.
# depth = 2 -> at most 4 branches, depth = 3 -> at most 8 branches

# 5. subsample -> fraction of training rows used for each tree, prevents overfitting
# subsample = 0.8 -> only use 80% of training data for each tree

# 6. colsample_bytree -> fraction of total predictors to use for each tree
# similar to max_features in random forest
# colsample_bytree = 0.8 -> use 80% of predictors, in our case we have 4 so use 3.

# 7. gamma -> gamma = 0 makes big complex trees, gamma = 5 makes small simple trees

# 8. min_child_weight -> number of samples or weights needed to make a split.
# similar to min_sample_split in random forest, BUT if we use weights, can be different
# min_child_weight = 5 -> need at least 5 pokemon to make a split
# for instance, if we give a legendary a weight of 5, and regular pokemon a weight of 1,
# 2 legendary pokemon and 5 regular pokemon -> (2*5) + 5 = 15 as weight.

# 9. scale_pos_weight -> do number_majority_class / number_rare_class
# this accounts for class imbalance! We have a lot more non legendaries than legendaries
# NOTE: This only works for binary classification!

# 10. random_state = 1 -> set the seed as 1, this makes results reproducible.
# think of this as a minecraft seed, the same minecraft seed always gives a same world!


# lets get our non bankrupt and bankrupt companies:

non_bankrupt_num = sum(y_train == 0)
print("Non Bankrupt companies: ", non_bankrupt_num)

bankrupt_num = sum(y_train == 1)
print("Bankrupt companies: ", bankrupt_num)


xgboost_model = XGBClassifier(n_estimators = 500,
                          learning_rate = 0.05, # eta
                          max_depth = 3,
                          subsample = 0.8,
                          colsample_bytree = 0.8,
                          gamma = 3,
                          min_child_weight = 10,
                          scale_pos_weight = (non_bankrupt_num/bankrupt_num),
                          random_state = 1
                          )



xgboost_model.fit(X_train, y_train)

y_preds = xgboost_model.predict(X_test)

results_df =  pd.DataFrame({"actual values":y_test,"predicted values":y_preds})
results_df

```

```python
from sklearn.metrics import average_precision_score

# lets not use y_preds, lets use probabilities, AKA y_probs!

# ALERT: using only y_preds makes one threshold, usually 0.5!
# This is because we only have 0 or 1!!!

# using y_probs gives multiple thresholds since we have multiple probability values!


y_probs = xgboost_model.predict_proba(X_test)[:, 1]
# [:,1] selects the probability of class 1

pr_auc_score = round(average_precision_score(y_test, y_probs),3)

print("XGBoost's PR-AUC score is:", pr_auc_score)

auc_check(pr_auc_score)
```

Interesting, XGBoost's CV PR-AUC score is 0.477, but XGBoost's testing set PR-AUC score is 0.456.

A PR-AUC score of 0.456 is still great though, and we can use this model to predict future bankrupt companies.


For fun, lets check the accuracy score.





```python
from sklearn.metrics import accuracy_score


accuracy_score = round(accuracy_score(y_test, y_preds),3)

print("The model's accuracy score is: ", accuracy_score)

```

Keep in mind that the accuracy score is not everything. A model can predict all companies as non-bankrupt and get a high accuracy score.

## üèÜ PR-AUC curve for XGBoost predictions

```python
from sklearn.metrics import precision_recall_curve


# lets not use y_preds, lets use probabilities, AKA y_probs!

y_probs = xgboost_model.predict_proba(X_test)[:, 1]
# [:,1] selects the probability of class 1


precision, recall, thresholds = precision_recall_curve(y_test, y_probs)

print(precision,recall,thresholds)



```

```python
fig = plt.figure(figsize = (15,8))

plt.plot(recall, precision, linewidth = 4, color = "green",
         label = "XGBoost PR-AUC curve")

plt.axhline(y = 0.03, color = "red", linewidth = 4,
            label = "Baseline PR-AUC curve")

# fill the area under the xgboost pr-auc curve:
plt.fill_between(recall, precision, color = "green", alpha = 0.2,
                 label = "AUC for XGBoost = 0.456")

# fill the area under the baseline pr-auc curve:
plt.fill_between(recall, 0, 0.03, alpha = 0.2, color = "red",
                 label = "AUC for baseline = 0.03")

plt.title("Precision Recall Curve for XGBoost", fontsize = 25)

plt.xlabel("Recall", fontsize = 15)

plt.ylabel("Precision", fontsize = 15)

plt.legend();

```

## How to read the Precision Recall Curve

A high curve in the top right part of the plot implies that there is both high precision and recall. This results in a bigger shaded area. We want models that have big areas since the area represents the PR-AUC score.

Conversely, a low curve in the bottom left part of the plot implies that there is both low precision and recall. This results in a lower shaded area.

Since the PR-AUC score was 0.456, it makes sense that we do not see a high or low curve.

We have evidence that our XGBoost model performs reasonably well at predicting bankrupt companies.  For context, the baseline precision for a model that predicts randomly is about 0.03 since about 3% of companies are bankrupt. A PR-AUC score of 0.456 is significantly higher than 0.03!
